```python
# Name: Syed Arham Naqvi
# Student Number: 100590852

import numpy as np
import datetime as dt

#Insertion Sort Algorithm
def insertionSort(arr):
    for j in range(1,len(arr)):
        
        key = arr[j]
        i = j - 1
        
        while( (i > 0) and (arr[i]) > key ):
            arr[i+1] = arr[i]
            i -= 1
        arr[i+1] = key

size = 1000
#Creating a random array
randArr = np.random.randint(1,size+1, size=(size,))
#Creating a sorted array
sortedArr = np.arange(1,size+1,1)
#Creating an array sorted in reverse order
reverseArr = np.arange(size, 0, -1)

#calculating and logging runtime for random array
start = dt.datetime.now()
insertionSort(randArr)
end = dt.datetime.now()
print(f"Random array size: {size} Insertion Sort runtime: {end-start}")

#calculating and logging runtime for sorted array
start = dt.datetime.now()
insertionSort(sortedArr)
end = dt.datetime.now()
print(f"Sorted array size: {size} Insertion Sort runtime: {end-start}")

#calculating and logging runtime for array sorted in reverse order
start = dt.datetime.now()
insertionSort(reverseArr)
end = dt.datetime.now()
print(f"Reverse sorted array size: {size} Insertion Sort runtime: {end-start}")
```

```python
#Mergesort

def mergesort(arr):
    
    if len(arr) <= 1:
        return
    
    middle = len(arr) // 2
    left_arr = arr[:middle]
    right_arr = arr[middle:]
    
    mergesort(left_arr)
    mergesort(right_arr)
    
    leftptr = 0
    rightptr = 0    
    for i in range(0,len(arr)):
        
        if leftptr >= len(left_arr):
            arr[i] = right_arr[rightptr]
            rightptr += 1
            continue
        elif rightptr >= len(right_arr):
            arr[i] = left_arr[leftptr]
            leftptr += 1
            continue
            
        if left_arr[leftptr] <= right_arr[rightptr]:
            arr[i] = left_arr[leftptr]
            leftptr += 1
        else:
            arr[i] = right_arr[rightptr]
            rightptr += 1


array = np.random.randint(0,10,6)
array = list(array)
print(array)
mergesort(array)
print(array)
```

```python
import matplotlib.pyplot as plt

# comparing and plotting the time complexities of the two sorting algorithms with increasing sizes of n

size_max = 10000000
experiments = 7
results = np.zeros((experiments,3))
exp_num = 0

for n in range(0,experiments):
    
    arraySize = size_max//(10**((experiments-1) - n))
    
    results[exp_num][0] = arraySize
    
    # array = list(np.random.randint(0,n,n))
    
    # start = dt.datetime.now()
    # insertionSort(array)
    # end = dt.datetime.now()
    # results[exp_num][1] = (end - start).microseconds

    array = list(np.random.randint(0,arraySize,arraySize))

    start = dt.datetime.utcnow()
    mergesort(array)
    end = dt.datetime.utcnow()
    results[exp_num][2] = (end - start).total_seconds()*100000
    
    exp_num += 1
plt.figure(figsize=(20,6))
# plt.plot(results[:,0], results[:,1], label="Insertion Sort", color="blue", marker="x")
plt.plot(results[:,0], results[:,2], label="Merge Sort", color="red")
plt.xlabel("Size of Array (n)")
plt.ylabel("Runtime (ms)")
plt.title("Merge Sort v Insertion Sort Runtimes with Random Arrays of Increasing Size")
```
